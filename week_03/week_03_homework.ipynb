9. (3 pts) On average are Japanese cars more fuel-efficient than US cars? Write a function that can handle this comparison.

```python
def compare_fuel_efficiency(origin_a, origin_b):
    mpg_a = mpg[origin == origin_a]
    mpg_b = mpg[origin == origin_b]
    
    avg_a = np.mean(mpg_a)
    avg_b = np.mean(mpg_b)
    
    if avg_a > avg_b:
        print(origin_a, "cars are more fuel-efficient on average.")
    else:
        print(origin_b, "cars are more fuel-efficient on average.")

compare_fuel_efficiency('Japan', 'US')
```

10. (3 pts) What is the correlation coefficient of mpg and weight? What is a correlation coefficient? What does the result indicate about the relationship between mpg and weight?

```python
corr_matrix = np.corrcoef(mpg, weight)
print("Correlation coefficient:", corr_matrix[0, 1])
```

**Answer:** A correlation coefficient measures how much two things are related from -1 to 1. The negative number means as a car gets heavier, its gas mileage goes down.

Extra credit (2 pts): Write a function to find the deepest level of nested brackets
Sample Input: `{"a":1,"b":2,"c":{"d":3,"e":4,"f":{"g":5,"h":6}}}`
Sample Output: There are 3 nested layers

```python
def find_layers(data):
    data_string = str(data)
    
    max_count = 0
    current_count = 0
    
    for char in data_string:
        if char == '{':
            current_count += 1
            if current_count > max_count:
                max_count = current_count
        elif char == '}':
            current_count -= 1
            
    return max_count

sample_input = {"a":1,"b":2,"c":{"d":3,"e":4,"f":{"g":5,"h":6}}}
layers = find_layers(sample_input)
print("There are", layers, "nested layers")
```
