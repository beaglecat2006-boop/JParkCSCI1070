9. (3 pts) On average are Japanese cars more fuel-efficient than US cars? Write a function that can handle this comparison.

```python
def compare_fuel_efficiency(origin_a, origin_b):
    mpg_a = mpg[origin == origin_a]
    mpg_b = mpg[origin == origin_b]
    
    avg_a = np.mean(mpg_a)
    avg_b = np.mean(mpg_b)
    
    if avg_a > avg_b:
        print(f"{origin_a} cars are more fuel-efficient on average.")
    else:
        print(f"{origin_b} cars are more fuel-efficient on average.")
        
    print(f"{origin_a} average MPG: {avg_a:.2f}")
    print(f"{origin_b} average MPG: {avg_b:.2f}")

compare_fuel_efficiency('Japan', 'US')
```

10. (3 pts) What is the correlation coefficient of mpg and weight? What is a correlation coefficient? What does the result indicate about the relationship between mpg and weight?

```python
corr = np.corrcoef(mpg, weight)[0, 1]
print(f"Correlation coefficient: {corr:.4f}\n")
print("Answer:")
print("A correlation coefficient is a numerical measure ranging from -1 to 1 that indicates the strength and direction of a linear relationship between two variables.")
print("The result of ~ -0.83 indicates a strong negative correlation, meaning that as a car's weight increases, its miles per gallon (fuel efficiency) heavily tends to decrease.")
```

Extra credit (2 pts): Write a function to find the deepest level of nested brackets
Sample Input: `{"a":1,"b":2,"c":{"d":3,"e":4,"f":{"g":5,"h":6}}}`
Sample Output: There are 3 nested layers

```python
def max_depth(data):
    if not isinstance(data, dict) or not data:
        return 0
    return 1 + max(max_depth(v) for v in data.values())

sample_input = {"a":1,"b":2,"c":{"d":3,"e":4,"f":{"g":5,"h":6}}}
layers = max_depth(sample_input)
print(f"There are {layers} nested layers")
```
