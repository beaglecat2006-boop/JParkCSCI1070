```python
def preprocess_data(df: pd.DataFrame) -> tuple:
    df = df.copy()
    
    df = df.dropna()
    
    X = df.drop("target", axis=1)
    y = df["target"]
    
    X = pd.get_dummies(X, drop_first=True)
    
    return X, y
```

```python
def split_data(X,y,test_size: float=0.2):
    from sklearn.model_selection import train_test_split
    return train_test_split(X,y,test_size=test_size, random_state=42)
```

```python
def run_linear_regression(X_train, X_test, y_train, y_test, feature_name: str):
    from sklearn.linear_model import LinearRegression
    from sklearn.metrics import accuracy_score
    
    model = LinearRegression()
    
    model.fit(X_train[[feature_name]], y_train)
    preds = model.predict(X_test[[feature_name]])
    
    preds = (preds > 0.5).astype(int)
    
    return accuracy_score(y_test, preds)
```

```python
def find_best_k(X_train, X_test, y_train, y_test, k_range=range(1,21)):
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.metrics import accuracy_score
    
    best_k = 1
    best_score = 0
    
    for k in k_range:
        model = KNeighborsClassifier(n_neighbors=k)
        model.fit(X_train, y_train)
        preds = model.predict(X_test)
        score = accuracy_score(y_test, preds)
        
        if score > best_score:
            best_score = score
            best_k = k
            
    return best_k, best_score
```

```python
#1

import pandas as pd
import numpy as np

def load_data(filepath="credit_approval.csv"):
    """Load the credit approval dataset."""
    df = pd.read_csv(filepath)
    print("Data loaded successfully.")
    print("First 5 rows:")
    print(df.head())
    print("Data shape:", df.shape)
    return df

def clean_data(df, target_col="Approved"):
    """Clean the dataset: handle missing values, convert target."""
    df = df.replace("?", np.nan)
    df = df.dropna()
    
    df[target_col] = df[target_col].apply(lambda x: 1 if x == '+' else 0)
    
    df = pd.get_dummies(df, drop_first=True)

    print("\nData cleaning complete.")
    print(f"New shape: {df.shape}")
    print("First 5 rows of cleaned data:")
    print(df.head()) # Shows first 5 rows and new dummy columns

```

```python
#2

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score

def univariate_linear_regression(df, feature, target="Approved"):
    X = df[[feature]]
    y = df[target]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    predictions = model.predict(X_test)
    
    predictions_class = [1 if p > 0.5 else 0 for p in predictions]
    
    mse = mean_squared_error(y_test, predictions)
    acc = accuracy_score(y_test, predictions_class)
    
    return mse, acc
```

```python
#3

from sklearn.neighbors import KNeighborsClassifier

def find_best_k(X_train, X_test, y_train, y_test, max_k=20):
    best_k = 1
    best_score = 0
    
    for k in range(1, max_k):
        model = KNeighborsClassifier(n_neighbors=k)
        model.fit(X_train, y_train)
        score = model.score(X_test, y_test)
        
        if score > best_score:
            best_score = score
            best_k = k
            
    return best_k, best_score
```

```python
#4

from sklearn.linear_model import LogisticRegression

def run_logistic_regression(df, target="Approved"):
    X = df.drop(columns=[target])
    y = df[target]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    model = LogisticRegression(max_iter=1000)
    model.fit(X_train, y_train)
    
    accuracy = model.score(X_test, y_test)
    
    return accuracy
```
