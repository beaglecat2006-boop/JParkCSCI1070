3. (3 pts) Create a function that checks if two words have the same characters. Make sure you use loops and proper data types to solve this. Dont use built in functions like sorted, etc.
<br /> example function call: same_characters(one="dormitory", two="dirty room")
<br /> example output: "'dormitory' and 'dirty room' have the same characters" <br />
<br /> example function call: same_characters(one="analytical", two="titanically")
<br /> example output: "'analyical' and 'titanically' have the same characters"
<br /> (Assume you'll always get strings and if you dont, return a message indicating the input is in the wrong formats) 

```python
def same_characters(str1, str2):
    if not isinstance(str1, str) or not isinstance(str2, str):
        return "Inputs must be strings"
    
    # Remove spaces
    one = str1.replace(" ", "")
    two = str2.replace(" ", "")
    
    # Count characters manually
    count_one = {}
    count_two = {}
    
    for char in one:
        count_one[char] = count_one.get(char, 0) + 1
    for char in two:
        count_two[char] = count_two.get(char, 0) + 1
    
    if count_one == count_two:
        return f"'{one}' and '{two}' have the same characters"
    else:
        return f"'{one}' and '{two}' do NOT have the same characters"

print(same_characters("dormitory", "dirty room"))
```

4. (3 pts) Create a function that prints a triangle of asterisks as below. 
The function accepts n as an integer and makes the longest leg of the triangle that length, decrementing by one until you close the triangle. The function also accepts a direction for the triangle as a variable. If you pass in "base-up" it returns the first, but "base-down" returns the second.
For example, for a triangle where n = 5, the following would be produced:

```python
#Base-up triangle
# * * * * *
# * * * *
# * * *
# * *
# *

#Base-down triangle
# *
# * *
# * * *
# * * * *
# * * * * *

def triangles(n, direction):
    if direction == "base-up":
        for i in range(n, 0, -1):
            print("* " * i)
    elif direction == "base-down":
        for i in range(1, n + 1):
            print("* " * i)
    else:
        print("Invalid direction")

triangles(5, "base-up")
triangles(5, "base-down")
```

9. (3 pts) Create a class called `Dice` that accepts a single argument called `number_of_dice`. 

Create an instance method called `roll`. 

`roll` should randomly pick a number for each die when called and print the results as a list. 

Create an instance of `Dice`. Then call `roll()` on your instance for each die. 

Hint: Inside your class create an empty list called `dice`. Use [randint](https://docs.python.org/3/library/random.html#random.randrange) function to randomly generate a function between 1 and 6

```python
# Helpful starter code
from random import randint

class Dice:
    def __init__(self, number_of_dice):
        self.number_of_dice = number_of_dice

    def roll(self):
        dice = []
        for _ in range(self.number_of_dice):
            dice.append(randint(1, 6))
        print(dice)

d = Dice(3)
d.roll()
```
