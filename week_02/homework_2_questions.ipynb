3. (3 pts) Create a function that checks if two words have the same characters. Make sure you use loops and proper data types to solve this. Dont use built in functions like sorted, etc.

```python
def same_characters(str1, str2):
    if not isinstance(str1, str) or not isinstance(str2, str):
        return "Inputs must be strings"
        
    one = str1.replace(" ", "").lower()
    two = str2.replace(" ", "").lower()
        
    count_one = {}
    count_two = {}
        
    for char in one:
        count_one[char] = count_one.get(char, 0) + 1
    for char in two:
        count_two[char] = count_two.get(char, 0) + 1
        
    if count_one == count_two:
        return f"'{str1}' and '{str2}' have the same characters"
    else:
        return f"'{str1}' and '{str2}' do NOT have the same characters"

print(same_characters("dormitory", "dirty room"))
```

4. (3 pts) Create a function that prints a triangle of asterisks as below. The function accepts n as an integer and makes the longest leg of the triangle that length, decrementing by one until you close the triangle. The function also accepts a direction for the triangle as a variable.

```python
def triangles(n, direction):
    if direction == "base-up":
        for i in range(n, 0, -1):
            print("* " * i)
    elif direction == "base-down":
        for i in range(1, n + 1):
            print("* " * i)
    else:
        print("Invalid direction")

triangles(5, "base-up")
triangles(5, "base-down")
```

5. (4 pts) Create a function that accepts a letter. Variables a, b, c, and d all start with a value of 0 and can each have a max value of 8. You need to find and return the value for each that results in the place that letter is in the alphabet.

```python
def letter_to_vars(letter):
    if not isinstance(letter, str) or len(letter) != 1 or not letter.isalpha():
        return "Invalid Input. Input must be a single alphabetic character."

    alphabet_pos = ord(letter.upper()) - ord('A') + 1
    a = min(alphabet_pos, 8)
    remaining = alphabet_pos - a
    b = min(remaining, 8)
    remaining -= b
    c = min(remaining, 8)
    remaining -= c
    d = remaining
    return f"a = {a}, b = {b}, c = {c}, and d = {d} for the letter {letter}"

print(letter_to_vars("J"))
```

6. (3 pts) Write a function that returns elements greater than 10 where both first and last digits of the number are odd.

```python
def odd_first_last(numbers):
    result = []
    for num in numbers:
        if num > 10:
            str_num = str(num)
            if int(str_num[0]) % 2 == 1 and int(str_num[-1]) % 2 == 1:
                result.append(num)
    return result

odd_first_last([3, 12, 15, 22, 27, 34, 37, 41, 53, 58, 63, 70, 89, 91, 100, 123])
```

8. (3 pts) Add a static method of your choice to the class below. Instantiate an object and call your method.

What is a static method? (1 pt for your answer here)

```python
class Actress:
    def __init__(self, name='Meryl Streep'):
        self.name = name
        
    def act(self, line):
        print(f'{self.name}: {line}')
        
    @staticmethod
    def favorite_movie():
        return "The Devil Wears Prada"

meryl = Actress()
meryl.act("This is my line...")
print(Actress.favorite_movie())
print("Definition: A static method is bound to the class rather than instances. It cannot access or modify instance/class state.")
```

9. (3 pts) Create a class called Dice that accepts a single argument called number_of_dice. Create an instance method called roll.

```python
from random import randint

class Dice:
    def __init__(self, number_of_dice):
        self.number_of_dice = number_of_dice
        
    def roll(self):
        dice = []
        for _ in range(self.number_of_dice):
            dice.append(randint(1, 6))
        print(dice)

d = Dice(3)
d.roll()
```

BONUS: (3 pts) Create a class that implements a versioned key-value store with rollback capability

```python
class VersionedKVStore:
    def __init__(self):
        self.current_version = 0
        self.history = {0: {}}
        
    def set(self, key, value):
        new_state = dict(self.history[self.current_version])
        new_state[key] = value
        self.current_version += 1
        self.history[self.current_version] = new_state
        
        keys_to_delete = [v for v in self.history if v > self.current_version]
        for v in keys_to_delete:
            del self.history[v]
            
    def get(self, key, version=None):
        if version is None:
            version = self.current_version
        if version not in self.history:
            return "Version does not exist"
        return self.history[version].get(key, "Key not found")
        
    def rollback(self, version):
        if version in self.history:
            self.current_version = version
        else:
            print("Invalid version for rollback")
            
    def get_history(self, key):
        key_history = []
        for v in sorted(self.history.keys()):
            if v == 0: continue
            state = self.history[v]
            if key in state and (v == 1 or self.history[v-1].get(key) != state[key]):
                key_history.append(f"Version {v}: {state[key]}")
        return key_history

kv = VersionedKVStore()
kv.set("username", "alice")
kv.set("username", "bob")
print(kv.get_history("username"))
```
