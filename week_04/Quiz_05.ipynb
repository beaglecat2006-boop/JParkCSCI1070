```python
from decimal import Decimal
import os
import pandas as pd

file_name = "JEOPARDY_DATA.csv"
if os.path.exists(file_name):
    df = pd.read_csv(file_name)
else:
    df = pd.DataFrame(
        {
            "Round": ["JEOPARDY!", "DOUBLE JEOPARDY!", "FINAL JEOPARDY!"],
            "Value": ["$200", "$1,000", "None"],
        }
    )

df.columns = df.columns.str.strip().str.lower()
df = pd.get_dummies(df, columns=["round"])

df["value"] = (
    df["value"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .replace("None", "0")  
    .map(Decimal)
)

print(f"Output type: {type(df['value'].iloc[0])}")

```
