```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Optional: make plots appear inline (Jupyter only)
%matplotlib inline
```

```python
# Load the survey CSV file
df = pd.read_csv('CSCI_Spring_2025_survey.csv')

# Quick look at the data
df.head()
```

```python
# Standardize the 'Place' column
df['Place'] = df['Place'].str.strip().str.lower()

place_mapping = {
    'ny': 'new york',
    'new york city': 'new york',
    'la': 'los angeles',
    'los angeles, ca': 'los angeles'
}

df['Place'] = df['Place'].replace(place_mapping)
df['Place'].head()
```

```python
df['Pets'] = df['Pets'].str.strip().str.lower()

pets_mapping = {
    'dog': 'dog',
    'dogs': 'dog',
    'puppy': 'dog',
    'cat': 'cat',
    'cats': 'cat',
    'kitten': 'cat',
    'none': 'none'
}

df['Pets'] = df['Pets'].replace(pets_mapping)
df['Pets'].head()
```

```python
df['Favorite_Food'] = df['Favorite_Food'].str.strip().str.lower()

food_mapping = {
    'pizza': 'pizza',
    'burger': 'burger',
    'hamburger': 'burger',
    'sushi': 'sushi',
    'pasta': 'pasta'
}

df['Favorite_Food'] = df['Favorite_Food'].replace(food_mapping)
df['Favorite_Food'].head()
```

```python
df['Major'] = df['Major'].str.strip().str.lower()

major_mapping = {
    'cs': 'computer science',
    'computer science': 'computer science',
    'math': 'mathematics',
    'mathematics': 'mathematics'
}

df['Major'] = df['Major'].replace(major_mapping)
df['Major'].head()
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')  # Convert to numbers, invalid to NaN
df['Age'].head()
```

```python
df['Pets'].value_counts().plot.pie(autopct='%1.1f%%', title='Class Pets Distribution')
plt.ylabel('')
plt.show()
```

```python
### Pets Distribution
Most students have dogs as pets, followed by cats and students with no pets.
```

```python
df['Major'].value_counts().plot.bar(title='Class Majors')
plt.xlabel('Major')
plt.ylabel('Number of Students')
plt.show()
```

```python
### Majors
The majority of students are computer science majors, followed by mathematics majors.
```
