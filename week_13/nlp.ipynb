# Stop Words

```python
# stop words are common words that have
# low information value.
# nltk.download('stopwords')
sw = set(nltk.corpus.stopwords.words('english'))
sw
```

```python
text_sentences = nltk.corpus.gutenberg.sents('carroll-alice.txt')[:10]
text_sentences
```

```python
# stop words are all lowercase 
# so we want to make sure all our words are
# lowercase when the process them
for sentence in text_sentences: 
    filtered_list = [word for word in sentence if word.lower() not in sw]
    print(filtered_list)
```

```python
#Google adopted word stemming in 2003
#It still does but it uses a more sophisticated
#Approach. Before this, search words like "fish"
#Would not give results for "fishing" or "fishes"

from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer(language='english')
word = 'flying'
stemmer.stem(word)
```

```python
#Lemmatization

#Lemmatization is finding a form of a word that is related in the dictionary
#The process for calculating "lemmas" is more complex but more accurate
#Lemmatization considers context AND converts the word
#to its "meaningful base form"(a lemma). Words can have multiple lemmas

import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
word = 'changing'
lemmatizer.lemmatize(word)
```

```python
nltk.download('vadar_lexicon')
analyzer = SentimentIntensityAnalyzer()
text = """
       Amazing
       """
sentiment = analyzer.polarity_scores(text)
print(sentiment)
```

```python
import pandas as pd
sentiments = []
compound = sentiment['compound']
pos = sentiment['pos']
neg = sentiment['neg']
neu=sentiment['neu']

sentiments.append({"text":text,"positive":pos,"negative":neg, "neutral":neu})

pd.DataFrame(sentiments)
```
