```python
import torch
import torch.nn as nn
import torch.optim as optim
import io

csv_data = """Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction,Age,Outcome
6,148,72,35,0,33.6,0.627,50,1
1,85,66,29,0,26.6,0.351,31,0
8,183,64,0,0,23.3,0.672,32,1
1,89,66,23,94,28.1,0.167,21,0
0,137,40,35,168,43.1,2.288,33,1
5,116,74,0,0,25.6,0.201,30,0""" 

import pandas as pd
df = pd.read_csv(io.StringIO(csv_data))
X = torch.tensor(df.iloc[:, :-1].values, dtype=torch.float32)
y = torch.tensor(df.iloc[:, -1].values, dtype=torch.float32).unsqueeze(1)

model = nn.Sequential(
    nn.Linear(8, 12),
    nn.ReLU(),
    nn.Linear(12, 1),
    nn.Sigmoid()
)
loss_fn = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

for epoch in range(100):
    optimizer.zero_grad()
    outputs = model(X)
    loss = loss_fn(outputs, y)
    loss.backward()
    optimizer.step()
```

```python
#Model trained with SGD optimizer successfully
#decreased loss over time(Did require more traning steps(epochs)
#to reach the same level as Adam)
#SGD performs worse than Adam(when using default learning rates)
#The small scale of the Diabetes dataset contains features with widly
#different ranges. Adam thrives better because its custom learnes rates quickly. 
```
