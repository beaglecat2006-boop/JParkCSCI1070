```python
import pandas as pd
import os
import glob
from io import BytesIO

belize_listings_path = "belize_airbnb_listings.csv"

# Try default path first, then search the repo, then prompt upload (widget) as fallback.
if os.path.exists(belize_listings_path):
	belize_df = pd.read_csv(belize_listings_path)
else:
	# search recursively for a CSV that looks like the dataset
	matches = glob.glob('**/*belize*airbnb*listings*.csv', recursive=True) + glob.glob('**/*belize*.csv', recursive=True)
	matches = list(dict.fromkeys(matches))  # dedupe
	if matches:
		belize_listings_path = matches[0]
		print(f"Found file at: {belize_listings_path}")
		belize_df = pd.read_csv(belize_listings_path)
	else:
		# fallback: show upload widget so user can provide the CSV in the notebook
		try:
			from ipywidgets import FileUpload
			from IPython.display import display
			uploader = FileUpload(accept='.csv', multiple=False)
			display(uploader)
			print("No local CSV found. Please upload 'belize_airbnb_listings.csv' with the widget above and then re-run this cell.")
			if uploader.value:
				uploaded = next(iter(uploader.value.values()))
				belize_df = pd.read_csv(BytesIO(uploaded['content']))
			else:
				# create empty placeholder to avoid NameError in later cells
				belize_df = pd.DataFrame()
		except Exception:
			# non-interactive fallback
			print("No file found and upload widget unavailable. Create or place 'belize_airbnb_listings.csv' in the working directory.")
			belize_df = pd.DataFrame()

# show dataframe (or empty placeholder)
belize_df
```

```python
neighborhood_collect = belize_df[['price', 'number_of_reviews', 'availability_365', 'neighbourhood']].groupby("neighbourhood").agg({
        "price": "mean",
        "number_of_reviews": "sum",
        "availability_365": "max"})

print(neighborhood_collect)
```
