```python
Forward and Backward Propagation

Forward--> Takes input values and propagates forward through the hidden layers to the output. Used to calculate predictions and errors. 

Backward--> Go back one layer at a time from output to input to see how small changes impact the difference between the predicted and actual values. 
Working to minimize the loss function. 
```
